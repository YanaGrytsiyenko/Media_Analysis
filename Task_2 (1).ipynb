```python
import re
import nltk
from nltk.corpus import stopwords

nltk.download('stopwords')
stop_words = set(stopwords.words('russian'))

# Функція для очищення
def clean_text(text):
    text = str(text).lower()
    text = re.sub(r'[^a-zа-яёїієґ0-9\s]', '', text)  # прибираємо пунктуацію
    text = re.sub(r'\d+', '', text)  # прибираємо цифри
    text = ' '.join(word for word in text.split() if word not in stop_words)
    return text

# Створюємо нову колонку для кластеризації
df['text_for_clustering'] = df['Заголовок'].apply(clean_text)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.cluster import KMeans

# Векторизація
vectorizer = TfidfVectorizer(max_features=1000)
X = vectorizer.fit_transform(df['text_for_clustering'])

# Кластеризація (наприклад, 5 кластерів)
k = 5
kmeans = KMeans(n_clusters=k, random_state=42)
df['cluster_label'] = kmeans.fit_predict(X)

# Подивимось приклади з кожного кластеру
for i in range(k):
    print(f"\n **Кластер** {i}")
    print(df[df['cluster_label'] == i]['Заголовок'].head(5).to_string(index=False))
```

**Кластер 0** – РФ, ринок, гуманітарна тематика
Матеріали, що стосуються теми виходу (або невиходу) Nestle з російського ринок, а також гуманітарні ініціативи, пов’язані з подарунками для дітей. Є відтінок геополітичного або репутаційного підтексту.

**Кластер 1** – Святкові дати, 24 липня, тематичні дні
Згадки приурочені до певних дат (24 липня, “день турботи про себе”), згадуються теми свят, традицій, офіційних приводів для згадок. Найбільш "календарний" кластер.

**Кластер 2** – Соціальні ініціативи та підтримка захисників
Публікації про подарунки військовим, соціальні ініціативи компанії, інклюзію. Є зміст про відповідальність бізнесу, підтримку, партнерства.

**Кластер 3** – Продукти, споживання
Огляди брендів, огляди морозива, споживчі рейтинги, огляди торгових мереж, автоматизація. Це найбільш "споживчий" та комерційно орієнтований кластер.

**Кластер 4** – HR, лідерство, розвиток
Згадки про Nestle як роботодавця, участь у HR-форумах, відзначення топ-менеджерів, аналітика з ринку праці. Позиціює бренд як частину професійної спільноти.

```python
# Фільтруємо останній тиждень
last_week = df[df['Дата виходу'] >= '2024-07-24']

# Підрахунок згадок по кожному кластеру
mentions_by_cluster = last_week['cluster_label'].value_counts().sort_index()

# Виведемо результат
print("Згадки кластерів за останній тиждень:")
for i, count in mentions_by_cluster.items():
    print(f"Кластер {i}: {count} згадок")
```

Цей підхід дозволив виділити основні групи інформаційних повідомлень у масиві та оцінити їхню активність у часі.

Найпоширенішим наративом за останній тиждень була "споживча" тематика (кластер 3), за нею — "святкова" тематика (кластер 1). Тут до речі, видно, що фігурує дата 24липня - а саме ця дата була пікова в аналізі завдання 1.
